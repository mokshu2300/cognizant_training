```python
dataJoin=pd.DataFrame({
    "ID":[1,2,3],
    "Name":["apple","banana","grapes"]})
dataJoin2=pd.DataFrame({
    "ID":[1,2,3],
    "Marks":[10,20,30]})
result=pd.merge(dataJoin,dataJoin2, on="ID")
print(result)
```
