```python
import numpy as np
products=["tv","smartwatch","airbuds","ipad"]
sales_2024=[100,800,500,700]
sales_2025=[300,900,400,200]
x=np.arange(len(products))
width=0.25
plt.bar(x-width/2,sales_2024,width)
plt.bar(x-width/2,sales_2025,width)
plt.show()
```

```python
import numpy as np
products=["smartphone","macbook","ipad","airbuds"]
offline_sales=[200,800,500,600]
online_sales=[100,500,400,300]
plt.bar(products,offline_sales)
plt.bar(products,online_sales,bottom=online_sales)
plt.show()
```
