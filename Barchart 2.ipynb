```python
import matplotlib.pyplot as plt
import numpy as np
products=["iphone","macbook","ipad","airbuds"]
sales_2023=[78000,95500,44500,88500]
sales_2024=[7500,50000,80000,45000]
sales_2025=[25000,5200,2800,1200]
sales_2026=[12000,60000,32000,23000]
x=np.arange(len(products))
width=0.35
plt.bar(x-width/3,sales_2023,width,label='2023')
plt.bar(x+width/3,sales_2024,width,label='2024')
plt.bar(x+width/2,sales_2025,width,label='2025')
plt.bar(x-width/2,sales_2026,width,label='2026')
plt.xlabel("products")
plt.title("Sales comparison: 2023 vs 2024 vs 2025 vs 2026")
plt.xticks(x,products)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
products=["iphone","macbook","ipad"]
online=[78000,95500,44500]
offline=[7500,50000,80000]
whole_sale=[25000,5200,2800]
plt.bar(products,online,label='online')
plt.bar(products,offline,bottom=online,label='offline')
bottom_value=[online[i]+offline[i] for i in range(len(products))]
plt.bar(products,whole_sale,bottom=bottom_value,label='whole_sale')                                                              
plt.xlabel("products")
plt.ylabel("revenue")
plt.title("revenue of the distributed sales")
plt.legend()
plt.xticks(x,products)
plt.show()
```
