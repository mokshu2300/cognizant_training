```python
def square(x):
    return x*x

print(square(9))
```

```python
square=lambda x:x*x
print(square(8))
```

```python
multi=lambda a,b:a*b
print(multi(6,8))
```

```python
def multi(a,b):
    return a*b

print(multi(5,9))
```

```python
check=lambda num:"Even" if num%2==0 else "odd"
print(check(8))
```

```python
students=[("suresh",25),("mahesh",35),("rajesh",45)]
result=sorted(students, key=lambda x:x[1])
print(result)
```

```python
numbers=[1,2,3,4,5]
result=list(map(lambda x: x*x, numbers))
print(result)
```

```python
a=[4,9,5,7,6]
result=list(filter(lambda num:num%2==0,a))
print(result)
```

```python
a=[4,9,5,7,6]
result=list(filter(lambda num:num%2==1,a))
print(result)
```

```python
from functools import reduce
numbers=[1,2,3,4,5]
result=reduce(lambda x,y: x+y,numbers)
print(result)
```
