```python
class Parent:
    def func1(self):
        print("This function belongs to parent class")

class child(Parent):
    def func1(self):
        print("This function belongs to child class")
object=child()
object.func1()
```

```python
class First:
    def func1(self):
        print("This function belongs is in first class")

class second(First):
    def func1(self):
        print("This function belongs is in second class")
object=second()
object.func1()
```
