```python
customers=list(range(1,100))
sample=customers[::10]
print("Convenience sample",sample)
```

```python
employee={
    "A":10,
    "B":20,
    "C":40,
    "D":50
}
sample=[name for name,exp in employee.items() if exp>=10]
print("Judgement sample: ", sample)
```

```python
networks={
    "A":["B","C"],
    "B":["D"],
    "C":["F"]
}
sample=["A"]
sample.extend(networks["A"])
sample.extend(networks["B"])
print("snowBall",sample)
```

```python
import pandas as pd
data={
    "customer_id":range(1,101),
    "city":["hyderabad"]*50+["pune"]*50
}
result=pd.DataFrame(data)
e=result.sample(n=10)
print(e)
```

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.plot([1,2,3],[10,20,30])
plt.subplot(1,2,2)
plt.plot([1,2,3],[30,20,10])
plt.show()
```

```python
months=["jan","feb","mar"]
sales=[200,300,400]
profit=[50,30,20]
plt.plot(months,sales,label="sales")
plt.plot(months,profit,label="profit")
plt.legend()
plt.show()
```

```python
months=["jan","feb","mar"]
sales=[200,300,400]
profit=[50,30,20]
plt.plot(months,sales,label="sales",alpha=0.5)
plt.plot(months,profit,label="profit",alpha=0.5)
plt.legend()
plt.show()
```
