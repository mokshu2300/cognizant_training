```python
customers=list(range(1,100))
k=10
sample=customers[::k]
print("Systematic sample:",sample)
```

```python
import random
developer=list(range(1,51))
tester=list(range(51,81))
manager=list(range(81,100))
sample_dev=random.sample(developer,5)
sample_test=random.sample(tester,6)
sample_mrg=random.sample(manager,2)
final_sample=sample_dev+sample_test+sample_mrg
print(final_sample)
```

```python
clusters={
    "hyderabad":[1,2,3,4,5],
     "pune":[15,14,12],
    "chennai":[25,23,26,14,12]
}
select_clusters=["hyderabad","pune"]
sample=[]
for cluster in select_clusters:
    sample.extend(clusters[cluster])
    print("cluster sample:",sample)
     
```
