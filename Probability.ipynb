```python
import random
toss=random.choice(["Head","Tail"])
print(toss)
```

```python
heads=0
tails=0
for i in range(1,100):
    toss=random.choice(["Head","Tail"])

    if toss=="Head":
        heads +=1
    else: 
        tails +=1
print("Head :",heads)
print("Tails :",tails)
```
