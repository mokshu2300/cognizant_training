```python
import numpy as np
arr1=([11,22,33,44])
arr2=([55,66,77,88])
arr3=np.concatenate((arr1,arr2))
print(arr3)
```

```python
arr=np.array([10,20,30,40,30,50])
x=np.where(arr==30)
print(x)
```

```python
arr=np.array([10,20,30,40,30,50])
x=np.sort(arr)
print(x)
```

```python
import numpy as np
marks=np.array([40,76,83,96,68,56])
print(marks[marks>=45])
```

```python
marks=np.average([40,76,83,96,68,56])
print(marks)
```

```python
marks1=np.min([40,76,83,96,68,56])
marks2=np.max([40,76,83,96,68,56])
marks3=np.mean([40,76,83,96,68,56])
print(marks1)
print(marks2)
print(marks3)
```
